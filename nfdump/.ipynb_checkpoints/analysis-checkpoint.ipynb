# Reading the csv into a pandas dataframe

```python
df = pd.read_csv("huge.csv", sep=',', header=0)

list(df.columns) #list columns
```

## Top 10 src IP addresses that send the most flows to port 22. 

```python
df_p_22 = df[df['dp'] == 22]

df_p_22
```

# Top 4 active source IP addresses to 22.Get the addresses with destiny port 22

```python
top_addr_22 = df_p_22['sa'].value_counts()

top_addr_22.head(10)
```

# TS:  the src IP that has generated the highest number of flows to port 22.

```python
flows_top_22 = top_addr_22[0]

address_top_22 = top_addr_22.index[0]

print("TS:\n")
print("Adress:") 
print(address_top_22) 
print()
print(" Numbre of flows:")
print(flows_top_22)
```

# How many destinations has TS contacted?

```python
TS = df.loc[df['sa'] == address_top_22]

da_ts = TS.groupby('da')['da'].nunique()

da_ts.count()
```

# Top 4 dst IP addresses which generated the most bytes and were contacted by TS - Get the rows

```python
# Addresses contacted by TS
addr_contacted_by_TS = da_ts.index 
# Rows with the da an address contacted by TS
gen_byte_TS = df.loc[df['da'].isin(addr_contacted_by_TS)]

ip_1 = gen_byte_TS.loc[gen_byte_TS['da'] =='195.212.40.29']
ip_2 = gen_byte_TS.loc[gen_byte_TS['da'] =='195.212.40.38']
ip_3 = gen_byte_TS.loc[gen_byte_TS['da'] =='195.212.40.1']
ip_4 = gen_byte_TS.loc[gen_byte_TS['da'] =='195.212.40.6']
```

```python
# HARD CODED 
ip_1 = gen_byte_TS.loc[gen_byte_TS['da'] =='195.212.40.29']
ip_2 = gen_byte_TS.loc[gen_byte_TS['da'] =='195.212.40.38']
ip_3 = gen_byte_TS.loc[gen_byte_TS['da'] =='195.212.40.1']
ip_4 = gen_byte_TS.loc[gen_byte_TS['da'] =='195.212.40.6']

#Sume
sum_ip_1 = ip_1['ipkt'].sum()
sum_ip_2 = ip_2['ipkt'].sum()
sum_ip_3 = ip_3['ipkt'].sum()
sum_ip_4 = ip_4['ipkt'].sum()

print("IP: " + "195.212.40.29" + "      " + "Count: ")
print(sum_ip_1)
print()

print("IP: " + "195.212.40.38" + "      " + "Count: ")
print(sum_ip_2)
print()

print("IP: " + "195.212.40.1" + "      " + "Count: ")
print(sum_ip_3)
print()

print("IP: " + "195.212.40.6" + "      " + "Count: ")
print(sum_ip_4)
print()
```
